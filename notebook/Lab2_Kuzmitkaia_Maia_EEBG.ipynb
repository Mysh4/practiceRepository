2) Creating seqMeasurer **(WE SHOULD NOT COUNT BLANK SPACES)**

```python
%%bash
cat > seqMeasurer
#!/bin/bash
grep "^>" IGHJ_Human.fa > lengths.txt
grep -v -h "^>" IGHJ_Human.fa | tr -d '\n' | wc -m >> lengths.txt
grep "^>" IGHJ_Mouse.fa >> lengths.txt
grep -v -h "^>" IGHJ_Mouse.fa | tr -d '\n' | wc -m >> lengths.txt
grep "^>" IGHJ_Zebrafish.fa >> lengths.txt
grep -v -h "^>" IGHJ_Zebrafish.fa | tr -d '\n' | wc -m >> lengths.txt

```

Verifying whether seqMeasurer is non-executable

Making script executable
